```python
with open(r,) as fp:
  data = json.load(fp)
```

```python
polygons = data['predictions'][0]['det_polygons']
```

```python
def get_x_y_w_h(polygon: list):
  """Get x, y, w, h from polygon

  Args:
      polygon (list): [x0, y0, x1, y1, ..., xn, yn]
  """
  x_pos = []
  y_pos = []
  for i, pos in enumerate(polygon):
    if i % 2 == 0:
      x_pos.append(pos)
    else:
      y_pos.append(pos)
  
  x = min(x_pos)
  y = min(y_pos)
  w = max(x_pos) - x
  h = max(y_pos) - y
  return dict(x=x, y=y, width=w, height=h)
```

```python
[get_x_y_w_h(polygon) for polygon in polygons]
```
