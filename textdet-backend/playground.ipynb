```python
def get_x_y_w_h(polygon: list):
  """Get x, y, w, h from polygon

  Args:
      polygon (list): [x0, y0, x1, y1, ..., xn, yn]
  """
  x_pos = []
  y_pos = []
  for i, pos in enumerate(polygon):
    if i % 2 == 0:
      x_pos.append(pos)
    else:
      y_pos.append(pos)
  
  x = min(x_pos)
  y = min(y_pos)
  w = max(x_pos) - x
  h = max(y_pos) - y
  return dict(x=x, y=y, width=w, height=h)
```

```python
import requests
import os
```

```python
def get_tasks(page=1, page_size=100):
  resp = requests.get('https://labelstudio.wevis.info/api/projects/1/tasks',
                     params={'page': page, 'page_size': page_size},
                     headers={'Authorization': 'Token e5f1ba3ecc0d6d5424d94f319fddc133dee895a5'})
  if resp.ok:
    return resp.json()
  return []

def delete_task(task_id):
  resp = requests.delete('https://labelstudio.wevis.info/api/tasks/%s' % task_id,
                     headers={'Authorization': 'Token e5f1ba3ecc0d6d5424d94f319fddc133dee895a5'})
  return resp.status_code == 204
```

```python
image_vote_count = os.listdir('../vote62/data/image/1/') # class 1 is the vote count paper

def delete_tasks(task_list):
    found = 0
    for task in task_list:
        try:
            image_name = task['data']['image'].split('/')[-1]
        except TypeError:
            continue
        # If image is not in vote count paper, delete it
        if image_name in image_vote_count:
            found += 1
        else:
            if delete_task(task['id']):
                print('Deleted %s' % task['id'])
            else:
                print('Cannot delete %s' % task['id'])
    return found
```

```python
page_i = 1
tasks = get_tasks(page=page_i)
while len(tasks) > 0:
    delete_tasks(tasks)
    print(f'Page {page_i} has {len(tasks)} tasks')
    page_i += 1
    tasks = get_tasks(page=page_i)
```
